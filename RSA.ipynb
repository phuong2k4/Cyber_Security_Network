```python
import random

def gcd(a,b):
  if b == 0:
    return a
  return gcd(b, a%b)

def is_prime(a):
  if a < 2:
    return False
  for i in range(2, int(a**0.5) + 1):
    if a%i==0:
      return False
  return True

def generator_Random_Primary(bits):
  while True:
    result = random.getrandbits(bits)
    # 1 << (bits - 1) make sure there is no bits missing
    result |= 1 << (bits - 1)
    # |= 1 check if at the end of bits is even or not
    result |= 1
    if is_prime(result):
      return result

def decryption_exponent_d(e, phi_n):
  for d in range(2, phi_n):
    if (e*d) % phi_n == 1:
      return d
  return -1

# def decryption_exponent_d(e, phi_n):
#   def egcd(a,b):
#     if a==0:
#       return (b,0,1)
#     g,y,x = egcd(b%a,a)
#     return (g, x-(b//a)*y,y)
#   g,x,y = egcd(e,phi_n)
#   if g!=1:
#     return None
#   return x % phi_n

def generate_keys(bits):
  p = generator_Random_Primary(bits)
  q = generator_Random_Primary(bits)
  print(p,q)
  while p == q:
    q = generator_Random_Primary(bits)
  n = p * q
  phi_n = (p-1) * (q-1)
  e = 0
  for i in range(2, phi_n):
    if gcd(i,phi_n) == 1 and is_prime(i):
      e = i
      break

  d = decryption_exponent_d(e, phi_n)
  public_keys = (e, n)
  private_keys = (d,n)
  return public_keys, private_keys

# def power(type_ed, type_keys, n):
#   return (type_ed ** type_keys) % n

def power(base, expo, m):
  res = 1
  base = base % m
  while expo > 0:
    if expo & 1:
      res = (res * base) % m
    base = (base * base) % m
    expo //= 2
  return res

def encrypt(M, e, n):
  return power(M,e,n)
def decrypt(C, d, n):
  return power(C,d,n)

puk, prk = generate_keys(14)

m = 123

e,n = puk
d,n = prk

c = encrypt(m,e,n);
m = decrypt(c,d,n)

print("Public keys: ", puk)
print("Private keys: ", prk)
print("Cipher Text: ", c)
print("Decypted: ", m)
```
