```python
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
from Crypto.Util.Padding import pad, unpad

pt = b'Cyber Security Network'
key = get_random_bytes(16)
iv = get_random_bytes(16)

cipher = AES.new(key, AES.MODE_CBC, iv = iv)
cipher_text = cipher.encrypt(pad(pt, AES.block_size))

print("Secret Key: ", key.hex().upper())
print("Cipher Text: ", cipher_text.hex().upper())

decipher = AES.new(key, AES.MODE_CBC, iv=iv)
decipher_text = unpad(decipher.decrypt(cipher_text), AES.block_size)

print(decipher_text)
```
